## 구매이력 데이터  

```python
import numpy as np
import pandas as pd
import datetime as datetime
import dataload
```

```python
buy_full = dataload.load_buy()
```

```python
buy_full.head()
```

```python
buy_full['year'] = buy_full['date'].apply(lambda x :str(x)[:4])
buy_2018 = buy_full[buy_full.year=='2018'][['date', 'sex', 'age', 'big_cat', 'sm_cat', 'qty']]
buy_2019 = buy_full[buy_full.year=='2019'][['date', 'sex', 'age', 'big_cat', 'sm_cat', 'qty']]
buy_2018['date'] = buy_2018['date'].apply(lambda x : pd.to_datetime(str(x), format='%Y%m%d'))
buy_2019['date'] = buy_2019['date'].apply(lambda x : pd.to_datetime(str(x), format='%Y%m%d'))
buy_2018.shape, buy_2019.shape
```

```python
buy_2018["date"] = buy_2018["date"] + datetime.timedelta(days=-1)
buy_2018["date"] = buy_2018["date"].replace(datetime.date(2017,12,31), datetime.date(2018,12,31))
```

```python
buy_2018_cat = buy_2018.groupby(['date', 'big_cat', 'sm_cat']).sum('qty').reset_index()[['date', 'big_cat', 'sm_cat', 'qty']]
buy_2019_cat = buy_2019.groupby(['date', 'big_cat', 'sm_cat']).sum('qty').reset_index()[['date', 'big_cat', 'sm_cat', 'qty']]
buy_2018_cat["date"] = buy_2018_cat["date"].apply(lambda x : str(x)[5:])
buy_2019_cat["date"] = buy_2019_cat["date"].apply(lambda x : str(x)[5:])
buy_2018_ts = pd.pivot_table(buy_2018_cat, index="sm_cat", columns="date", values="qty")
buy_2019_ts = pd.pivot_table(buy_2019_cat, index="sm_cat", columns="date", values="qty")
buy_2018_ts.shape, buy_2019_ts.shape
```

```python
buy_2018_ts.head()
```

```python
buy_2019_ts.head()
```

```python
buy_2018_ts = buy_2018_ts.fillna(0)
buy_2019_ts = buy_2019_ts.fillna(0)
```

```python
buy_ts = (buy_2018_ts + buy_2019_ts)/2
buy_ts.shape
```
