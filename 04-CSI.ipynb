## 1. 서울 소비자 심리지수 계산

```python
df_seoul_past = df_seoul_1[df_seoul_1['분류코드별']=='서울'].reset_index(drop=True)
df_seoul_2020 = df_seoul_2[df_seoul_2['분류코드별']=='서울'].reset_index(drop=True)
```

```python
print(df_seoul_past.shape)
print(df_seoul_2020.shape)
```

```python
df_seoul = pd.merge(df_seoul_past, df_seoul_2020, how = 'left', on = ['지수코드별','분류코드별'])
```

```python
print(df_seoul.shape)
```

```python
df_seoul.head(2)
```

```python
df_seoul['지수코드별'].unique()
```

```python
df_csi = df_seoul[df_seoul['지수코드별'].apply(lambda x: x in ["현재생활형편CSI","생활형편전망CSI","현재경기판단CSI","향후경기전망CSI","소비지출전망CSI","가계수입전망CSI"])].reset_index(drop=True)
```

```python
df_csi.head(2)
```

```python
df_csi.drop(['단위','Unnamed: 28'], axis=1, inplace=True)
```

```python
df_csi_seoul = df_csi.mean().tolist()
```

## 2. 서울 이외의 지역

```python
df_others_1.head(2)
```

```python
df_others_2.head(2)
```

```python
df_others = pd.merge(df_others_1, df_others_2, how = 'left', on = ['CSI_CODE별','BRANCH_CODE별'])
```

```python
df_others_csi = df_others[df_others['CSI_CODE별'].apply(lambda x: x in ["현재생활형편CSI","생활형편전망CSI","현재경기판단CSI","향후경기전망CSI","소비지출전망CSI","가계수입전망CSI"])].reset_index(drop=True)
```

```python
df_others_csi.head(2)
```

```python
df_group = df_others_csi.groupby(['BRANCH_CODE별']).mean().reset_index()
```

## 3. 데이터 병합

```python
df_group.head(2)
```

```python
df_csi_seoul = ["서울"] + df_csi_seoul
```

```python
df_final = df_group.append(pd.DataFrame([df_csi_seoul], columns = df_group.columns)).reset_index(drop=True)
```

```python
df_final.rename(columns={"BRANCH_CODE별":"시도"},inplace=True)
```
